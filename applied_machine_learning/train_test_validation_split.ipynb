```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

dir = os.path.dirname(os.path.realpath(__name__))
file = os.path.join(dir, 'titanic_cleaned.csv')

titanic = pd.read_csv(file, encoding='utf-8', delimiter=',')
titanic.head(10)
```

### Split into Train, Validation and Test Set

```python
from cProfile import label


features = titanic.drop('Survived', axis=1)
labels = titanic['Survived']

# Features [X] and Labels [Y] will have 60% of train and 40% rest
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.4, random_state=42)

# Validation and Test sets will both take 50% of the test data set
X_val, X_test, y_val, y_test = train_test_split(
    X_test, y_test, test_size=0.5, random_state=42)

# Confirm split
for dataset in [y_train, y_val, y_test]:
    print(round(len(dataset) / len(labels), 2))
```
